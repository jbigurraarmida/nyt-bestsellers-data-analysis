# Clean publishers process
This notebook will detail the cleaning process for the `publisher` column in the database for later analysis.

## Libraries
Here, we import the libraries needed for reading the data from the `.csv` file and for the cleaning process.

```python
import pandas as pd
import re
```

## Loading Dataset

```python
df = pd.read_csv('nyt_bestsellers_author_clean_publishers_2.csv')
```

```python
len(df)
```

## Initial normalization process

```python
def normalize_publisher(name):
  """
  This normalizes publisher's names, changing the string into
  upper case, and changing the acronyms:
  W. W. NORTON & CO -> W W NORTON & CO
  """
  if pd.isna(name):
    return name

  name = name.upper()

  name = re.sub(r'\.', '', name)

  name = re.sub(r'\s+', ' ', name).strip()

  return name
```

```python
df['publisher'] = df['publisher'].apply(normalize_publisher)
```

## Exploring inconsistencies

```python
"""
Show the unique values (sorted) from the `publisher` column
for further cleaning
"""
publishers = df['publisher'].sort_values().unique()
print(publishers)
```

```python
len(publishers)
```

```python
"""
Search for specific substring in publishers
Example: HARPER -> HARPERCOLLINS, HARPERONE, HARPERTEEN, ...

This way, we find instances we can replace later, to avoid false different
publishers, like:
- TOR
- TOR/TOM DOHERTY
- TOR BOOKS

Another way way this particular cell may help, is finding if there are book
formats in the publisher column that we can delete later to avoid noise in
our data. For example:
- ST. MARTIN’S, HARDCOVER AND PAPER
"""
df.loc[df['publisher'].str.contains("KODANSHA", na=False), 'publisher'].unique()
```

## Manual replacements

```python
"""
Replace particular instances in 'publisher' to clean the data. It searches
a whole string and replaces it.
"""
df['publisher'] = df['publisher'].replace("VERTICAL INC", "VERTICAL")
```

```python
"""
Replace substring, not whole cell
Particularly useful for instances such as:
- ST. MARTIN'S, HARDCOVER AND PAPER -> ST. MARTIN'S
And other instances having the same structure
"""
df['publisher'] = df['publisher'].str.replace(", HARDCOVER ONLY", "", regex=False)
```
